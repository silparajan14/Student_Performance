# **Student_Performance**

# **Description:**

The Student Performance Dataset is a dataset designed to examine the factors influencing academic student performance. The dataset consists of 10,000 student records, with each record containing information about various predictors and a performance index.

**Variables:**

*1.Hours Studied:*  The total number of hours spent studying by each student.

*2.Previous Scores:*  The scores obtained by students in previous tests.

*3.Extracurricular Activities:*  Whether the student participates in extracurricular activities (Yes or No).

*4.Sleep Hours:*  The average number of hours of sleep the student had per day.

*5.Sample Question Papers Practiced:*  The number of sample question papers the student practiced.

**Target Variable:**

*6.Performance Index:*  A measure of the overall performance of each student. The performance index represents the student's academic performance and has been rounded to the nearest integer. The index ranges from 10 to 100, with higher values indicating better performance.


**AIM : The dataset aims to provide insights into the relationship between the predictor variables and the performance index.**

**Now converting 'Extracurricular Activities' column to numerical format**

```python
df['Extracurricular Activities'].value_counts()
```

```python
#to convert categorical variables in "Extracurricular Activities" column into numerical format we are using LabelEncoder :
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Extracurricular Activities'] = le.fit_transform(df['Extracurricular Activities'])
```

**Splitting dataset :**

```python
X = df.iloc[:,:-1]
X
```

```python
y = df.iloc[:,-1]
y
```

**Applying StandardScaler and then Train Test Split :**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 1)
```

```python
#checking shapes of each train & test sets :
print("Shape of X train : ",X_train.shape)
print("Shape of X test : ",X_test.shape)
print("Shape of y train : ",y_train.shape)
print("Shape of y test : ",y_test.shape)
```

Importing Data Models :

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge,Lasso

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import numpy as np
```

**Creating objects and applying for loop to get different measures and to check accuracy :**

```python
knn = KNeighborsRegressor()
lr =LinearRegression()
dt = DecisionTreeRegressor()
rfr = RandomForestRegressor()
ridgeReg = Ridge(alpha=10)
lasso = Lasso(alpha = 0.5)

models = [knn,lr, dt,rfr,ridgeReg,lasso]

for i in models:
  print(i)
  i.fit(X_train,y_train)
  y_pred = i.predict(X_test)
  print("R2_score : ",r2_score(y_test,y_pred)*100)
  print("MeanAbsoluteError : ",mean_absolute_error(y_test,y_pred))
  print("MeanSquaredError : ",mean_squared_error(y_test,y_pred))
  print("RootMeanSquaredError : ",np.sqrt(mean_squared_error(y_test,y_pred)))
  print("_____________________________________________________________________")
```

DataFrame of 'Predicted_Performance_Index'	and 'Actual_Performance_Index'

```python
Performance_Index_df = pd.DataFrame({ "Predicted_Performance_Index" :np.rint(ridgeReg.predict(X_test)), "Actual_Performance_Index" : y_test }).reset_index(drop = True)
Performance_Index_df
```

**Plotting the predicted and original values :**

```python
for i in models:
  print(i)
  i.fit(X_train,y_train)
  y_pred = i.predict(X_test)
  print("R2_score : ",r2_score(y_test,y_pred)*100)
  print("_____________________________________________________________________")
  n = 50
  plt.figure(figsize=(12,6))
  plt.title(i)
  sns.scatterplot(pd.DataFrame({ "Prediction": y_pred[:n], "True": y_test[:n] }))
  plt.show()
  print("______________________________________________________________________________________________________________________________________________")
```

```python
models = [lr,ridgeReg,lasso]

for i in models:
  print(i)
  print("Slope of ",i, "is : ",lr.coef_)
  print("Y Intercept of ",i,"is : ",lr.intercept_)
  print('____________________________________________')
```
